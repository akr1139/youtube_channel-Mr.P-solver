```python
import sympy as sp
import numpy as np
```

$f(x)=e^{-a \sin \left(x^2\right)} \cdot \sin \left(b^x\right) \cdot \ln \left(c \sin ^2(x) / x\right)$

```python
x,a,b,c=sp.symbols('x a b c',real=True)
x**2-6*x+15
```

```python
f=sp.exp(-a*sp.sin(x**2))*sp.sin(b**x)*sp.log(c*sp.sin(x)**2/x)
f
```

```python
dfdx=sp.diff(f,x)
dfdx
```

```python
d4fdx4=sp.diff(f,x,4)
d4fdx4
```

$d 4 f d x 4 \cdot \operatorname{subs}([(x, 4),(a, 1),(b, 2),(c, 3)]) \cdot \operatorname{evalf}()$

```python
d4fdx4.subs([(x,4),(a,1),(b,2),(c,3)])
```

```python
d4fdx4.subs([(x,4),(a,1),(b,2),(c,3)]).evalf()
```

```python
d4fdx4_f=sp.lambdify((x,a,b,c),d4fdx4)
d4fdx4_f
```

```python
x=np.linspace(1,2,100)
x
```

```python
y=d4fdx4_f(x,a=1,b=2,c=3)
y
```

```python
import matplotlib.pyplot as plt

plt.plot(x,y)
plt.ylabel('$d^4 f/ dx^4$',fontsize=15)
plt.xlabel('$x$',fontsize=15)
```

```python
filt = np.ones(15)/15
```

```python
y_smooth = np.convolve(y, filt, mode='valid')
dysdx = np.gradient(y_smooth, x[7:-7])
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,3))
ax[0].plot(x,y, label='$y(x)$')
ax[0].plot(x[7:-7], y_smooth, label=r'$y_{{smooth}}(x)$')
ax[1].plot(x,dydx, label='$y\'(x)$', color='r')
ax[1].plot(x[7:-7],dysdx, label='$y_{smooth}\'(x)$', color='purple')
ax[1].set_ylim(-100,120)
ax[1].grid()
[a.legend() for a in ax]
[a.set_xlabel('Time [Days]') for a in ax]
ax[0].set_ylabel('Cases per Day')
ax[1].set_ylabel('$\Delta$ (Cases per Day) / $\Delta t$')
fig.tight_layout()
plt.show()
```
